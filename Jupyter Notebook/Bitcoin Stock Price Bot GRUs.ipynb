```python
def plot_bar(df,year):
    plt.figure(figsize=(10, 5))
    bar_width = 0.4  # Adjust the width of the bars as needed
    plt.bar(df['Date'].dt.month_name(), df['Open'], label='Open', width=bar_width,color='lightsalmon', align='edge')
    plt.bar(df['Date'].dt.month_name(), df['Close'],label='Close', width=bar_width, color='black',align='center')

    # Adding labels and title
    plt.xlabel('Month')
    plt.ylabel('Price')
    plt.title(f'Open and Close Prices by Month in {year}')

    # Displaying legend
    plt.legend()

    # Rotating x-axis labels for better readability
    plt.xticks(rotation=45)

    # Display the plot
    plt.show()
```

### Data Preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler(feature_range=(0,1))
```

```python
df1 = df.loc[(df['Date'] >= '2016-01-01') & (df['Date'] <= '2022-02-01')]
```

```python
df1 = df1.drop('Date',axis=1)
```

```python
df1.columns
```

```python
mms.fit(df1)
df2 = mms.fit_transform(df1)
df2 = pd.DataFrame(data=df2,columns=df1.columns)
```

```python
closed_df = df2['Close']
```

```python
def create_dataset(dataset, time_step=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-time_step-1):
        a = dataset[i:(i+time_step), 0]   ###i=0, 0,1,2,3-----99   100 
        dataX.append(a)
        dataY.append(dataset[i + time_step, 0])
    return np.array(dataX), np.array(dataY)
```

```python
cls_df = np.array(df2['Close']).reshape(2224,1)
```

```python
x,y = create_dataset(cls_df,time_step=15)
```

```python
x.shape,y.shape
```

```python
x = x.reshape(2208,15,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,y_train.shape
```

### Model Architecture

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense,LSTM,GRU,Dropout
```

```python
model = tf.keras.Sequential()
model.add(GRU(units=20,input_shape=(None,1),activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mean_squared_error')
model.summary()
```

```python
hist = model.fit(x_train,y_train,batch_size=32,epochs=10,validation_data=(x_test,y_test))
```

```python
sns.lineplot(hist.history['loss'],label='Training Loss')
sns.lineplot(hist.history['val_loss'],label='Validation Loss')
```

```python
pred_x = model.predict(x_test)
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(pred_x,label='Prediction')
sns.lineplot(y_test,label='Y-test')
```
